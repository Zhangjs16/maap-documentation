```python
import warnings
warnings.filterwarnings('ignore')
from maap.maap import MAAP
maap = MAAP()
import os
import glob
import geopandas as gpd
from shapely.geometry import Point
import h5py
```

```python
%%time

granules = maap.searchGranule(short_name='ATL08', version='003', bounding_box="-77,-26,-73,0", limit=600)
for g in granules:
    g.getLocalPath('files/') # You can also add the option overwrite=True/False. Defaults to False.

print(f"Downloaded all {len(granules)} files")
files = glob.glob('files/*')
limit = 1000
points = []
elevation_low = []
height_above_ground = []
for idx in range(len(files)):
    try:
        f = h5py.File(files[idx], 'r')
        beams = ['gt1l', 'gt1r', 'gt2l', 'gt2r', 'gt3l', 'gt3r']
        for beam in beams:
            x, y, z = f[f"{beam}/land_segments/longitude"][:], f[f"{beam}/land_segments/latitude"][:], f[f"{beam}/land_segments/dem_h"][:]
            points.extend(list(map(Point, zip(x, y, z))))
            elevation_low.extend(f[f"{beam}/land_segments/terrain/h_te_best_fit"][:])
            try:
                height_above_ground.extend(f[f"{beam}/land_segments/canopy"][:])
            except:
                height_above_ground.extend([None] * len(x))
    except Exception as e:
        pass
        # print(f"Skipping {files[idx]} because {e}")

df = gpd.GeoDataFrame({'geometry': points[0:limit], 'ElevationLow': elevation_low[0:limit], 'HeightAboveGround': height_above_ground[0:limit]}, crs='epsg:4326')
subset = df.cx[-77:-73, -26:0]
```
